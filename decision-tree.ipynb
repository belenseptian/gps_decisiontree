```python
import pandas as pd
```

```python
data = pd.read_csv('dataset.csv')
data.head(10)
```

```python
pd.isna(data)
```

```python
data.info()
```

```python
inputs = data.drop(columns='status')
inputs.head(10)
```

```python
output = data['status']
output.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(inputs, output, test_size=0.3)
tree_data = DecisionTreeClassifier()
tree_data.fit(xtrain,ytrain)
```

```python
score = tree_data.score(xtest,ytest)
score
```

```python
prediction = tree_data.predict([[5,104.764160,-2.913836,1.78,8,104.764183,-2.913868,0.94,4.43,]])
prediction
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree_data, out_file="tree_gps.dot", class_names=["1", "0"], feature_names=inputs.columns, impurity=True, filled=True)
```

```python
import graphviz
with open("tree_gps.dot") as fig:
    dot_graph = fig.read()
graphviz.Source(dot_graph)
```
